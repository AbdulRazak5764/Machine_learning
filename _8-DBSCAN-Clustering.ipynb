# DBSCAN CLUSTERING 
Density- Based Spatial Clustering

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_moons
```

```python
x,y=make_moons(n_samples=250,noise=0.05)
```

```python
x[:,0]
```

```python
df={"data1":[x[:,0]],"data2":[x[:,1]],"output":y}
```

```python
df={"data1":x[:,0],"data2":x[:,1]}
data=pd.DataFrame(df)
data.head()
```

```python
# To check if the data is Linear or Non linear\
sns.scatterplot(x="data1",y="data2",data=data)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
db=DBSCAN(eps=0.2,min_samples=5)
db.fit_predict(data)
```

```python
data["predict"]=db.fit_predict(data)
data
```

```python
sns.scatterplot(x="data1",y="data2",data=data,hue="predict")
plt.show()
```

```python
sns.scatterplot(x="data1",y="data2",data=data,hue=y)
plt.show()
```
