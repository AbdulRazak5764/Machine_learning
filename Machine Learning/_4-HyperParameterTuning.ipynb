# Hyper Parameter Tuning

```python
import pandas as pd
```

```python
data=pd.read_csv(r"C:\Users\Shaik.AbdulRazak\OneDrive\Documents\Raw data\polynomial_dataset_like_example.csv")
data.head()
```

```python
x=data.iloc[:,:-1]
y=data["Salary"]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
dt.fit(x_train,y_train)
```

```python
dt.score(x_test,y_test)*100
```

```python
dt.score(x_train,y_train)*100
```

```python
from sklearn.tree import DecisionTreeRegressor
dt1=DecisionTreeRegressor( criterion="squared_error",splitter="best",max_depth=10)
dt1.fit(x_train,y_train)
```

```python
dt1.score(x_test,y_test)*100
```

```python
dt1.score(x_train,y_train)*100
```

# GridSearch-Cv

```python
from sklearn.model_selection import GridSearchCV
```

```python
df={"criterion":["Squared_error","friedman_mse","absolute_error","prisson"],
    "splitter":["best","random"],
    "max_depth":[i for i in range(2,20)]}
df
```

```python
gd=GridSearchCV(DecisionTreeRegressor(),param_grid=df)
gd.fit(x_train,y_train)
```

```python
gd.best_params_
```

```python
gd.best_score_*100
```

# RandomizedSearch-Cv

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
rd=RandomizedSearchCV(DecisionTreeRegressor(),param_distributions=df,n_iter=20)
rd.fit(x_train,y_train)
```

```python
rd.best_params_
```

```python
rd.best_score_*100
```
