# Feature Scaling

```python
# Feature Scaling is two types:
# 1) Standardization
# 2) Normalization
```

# 1.Standardization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv(r"C:\Users\Shaik.AbdulRazak\OneDrive\Documents\Raw data\loan.csv")
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data["ApplicantIncome"].fillna(data["ApplicantIncome"].mean(), inplace=True)
data
```

```python
sns.distplot(data["ApplicantIncome"])
plt.show()
```

```python
data.describe()
```

```python
# from sklearn.preprocessing  we can import StandardScaler
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
ss.fit(data[["ApplicantIncome"]])
```

```python
ss.transform(data[["ApplicantIncome"]])
```

```python
data["ApplicantIncome_ss"]=pd.DataFrame(ss.transform(data[["ApplicantIncome"]]),columns=["x"])
data
```

```python
data.describe()
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.distplot(data["ApplicantIncome"])

plt.subplot(1, 2, 2)
sns.distplot(data["ApplicantIncome_ss"])
plt.show()
```

# 2.Normalization

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data1=pd.read_csv(r"C:\Users\Shaik.AbdulRazak\OneDrive\Documents\Raw data\loan.csv")
data1.head()
```

```python
data1.isnull().sum()
```

```python
data1.describe()
```

```python
sns.distplot(data1["ApplicantIncome"])
plt.show()
```

```python
# from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
ms=MinMaxScaler()
ms.fit(data1[["CoapplicantIncome"]])
ms.transform(data1[["CoapplicantIncome"]])
```

```python
data1["CoapplicantIncome_Min"] = ms.transform(data1[["CoapplicantIncome"]])
```

```python
data1.head()
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.title("Before")
sns.displot(data1["CoapplicantIncome"])

plt.subplot(1, 2, 2)
plt.title("After")
sns.distplot(data1["CoapplicantIncome_Min"])
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.title("Befor")
sns.distplot(data1["CoapplicantIncome"])

plt.subplot(1, 2, 2)
plt.title("After")
sns.distplot(data1["CoapplicantIncome_Min"])
plt.show()
```
