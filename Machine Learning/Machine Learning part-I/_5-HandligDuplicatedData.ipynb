# Function Transformer

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data2=pd.read_csv(r"C:\Users\Shaik.AbdulRazak\OneDrive\Documents\Raw data\loan.csv")
data2.head()
```

```python
data2.shape
```

```python
data2.isnull().sum()
```

```python
sns.distplot(data2["CoapplicantIncome"])
plt.show()
```

```python
# by using IQR method to remove the outlier
q3 = data2["CoapplicantIncome"].quantile(0.25) 
q4 = data2["CoapplicantIncome"].quantile(0.75)
iqr = q4-q3
```

```python
min_r=q3-(1.5*iqr)
max_r=q4+(1.5*iqr)
min_r
max_r
```

```python
data2=data2[data2["CoapplicantIncome"]<=max_r]
```

```python
sns.distplot(data2["CoapplicantIncome"])
plt.show()
```

```python
# by using function Transformer we cn take sklearn.preprocessing
from sklearn.preprocessing import FunctionTransformer
```

```python
ft=FunctionTransformer(func=np.log1p)
```

```python
ft.fit(data2[["CoapplicantIncome"]])
```

```python
ft.transform(data2[["CoapplicantIncome"]])
```

```python
data2["CoapplicantIncome_tf"]= ft.transform(data2[["CoapplicantIncome"]])
data2.head()
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.distplot(data2["CoapplicantIncome"])
plt.title("Before")

plt.subplot(1, 2, 2)
sns.distplot(data2["CoapplicantIncome_tf"])
plt.title("After")
plt.show()
```
