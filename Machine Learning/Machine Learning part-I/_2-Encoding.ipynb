# ORDINAL ENCODING

```python
import pandas as pd
```

```python
df=pd.DataFrame({"size":["s","m","l","xl","s","m","l","s","s","l","xl","m"]})
df
```

```python
ord_data=[["s","m","l","xl"]]
# from sklearn.prepoceesing import OrdinalEncoder
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=ord_data)
oe.fit(df[["size"]])
```

```python
oe.transform(df[["size"]])
```

```python
df["size_en"]=oe.transform(df[["size"]])
df
```

```python
# we also define the values also waht u want
ord_data1 = {"s": 0, "m": 1, "l": 2, "xl": 3}
df["size_en"] = df["size"].map(ord_data1)
print(df)
```

```python
df = pd.DataFrame({"size": ["s", "m", "l", "xl", "s", "m"]})
ord_data1 = {"s": 57, "m": 64, "l": 72, "xl": 98} # u give the values waht you want
df["size_encoded"] = df["size"].map(ord_data1)
print(df)
```
