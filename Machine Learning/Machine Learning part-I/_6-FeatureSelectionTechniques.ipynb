# Feature Selection Techniques (Forward and Backward Elimination)

```python
import pandas as pd
```

```python
#pip install mlxtend
```

```python
# from mlxtend.feature_selection we import the SequentialFeatureSelector
from mlxtend.feature_selection import SequentialFeatureSelector
```

```python
data=pd.read_csv(r"C:\Users\Shaik.AbdulRazak\OneDrive\Documents\Raw data\diabetes.csv")
data.head()
```

```python
x=data.iloc[:,:-1]
y=data["Outcome"]
```

```python
x.shape
```

```python
# from sklearn.linear_model we import the LogisticRegression
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
fs=SequentialFeatureSelector(lr,k_features=5,forward=True) # forward = false for backward selection
fs.fit(x, y)
```

```python
fs.feature_names
```

```python
fs.k_feature_names_
```

```python
fs.k_score_
```
