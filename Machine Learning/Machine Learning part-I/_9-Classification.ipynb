# <h2>1.Logistick Regression & (Binary Classification)-->Single Input </h2>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv(r"C:\Users\Shaik.AbdulRazak\OneDrive\Documents\Raw data\Social_Network_Ads.csv")
data.head()
```

```python
data.drop(columns=["EstimatedSalary"], inplace=True)
```

```python
data.head()
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="Age", y="Purchased", data=data)
plt.show()
```

```python
x=data[["Age"]]
y=data["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
lr.predict([[30]])
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Age", y="Purchased", hue="Purchased", data=data)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
import numpy as np

# Scatter plot (actual data)
sns.scatterplot(x="Age", y="Purchased", hue="Purchased", data=data)

# Predictions need sorted x-values for a proper regression line
x_sorted = np.sort(data["Age"])
y_pred = lr.predict(x_sorted.reshape(-1, 1))

# Plot regression line
plt.plot(x_sorted, y_pred, color="red")

plt.show()
```

# <h2>2.Logistick Regression & (Binary Classification)-->Multiple Input </h2>

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data2=pd.read_csv(r"C:\Users\Shaik.AbdulRazak\OneDrive\Documents\Raw data\placement.csv")
data2.head()
```

```python
# to check the data is linear separation or not
plt.figure(figsize=(6,4))
sns.scatterplot(x="cgpa",y="score",data=data2,hue="placed")
plt.show()
```

```python
x=data2.iloc[:,:-1]
y=data2["placed"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
lr.predict([[6.8,5.26]])
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
#pip install mlxtend
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=lr)
plt.show()
```

# <h2>4.Logistick Regression (Multiclass - Classification)</h2>

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data4=pd.read_csv(r"C:\Users\Shaik.AbdulRazak\OneDrive\Documents\Raw data\Iris.csv")
data4.head()
```

```python
data4["Species"].unique()
```

```python
sns.pairplot(data=data4)
plt.show()
```

```python
sns.pairplot(data=data4,hue="Species")
plt.show()
```

```python
x=data4.iloc[:,:-1]
y=data4["Species"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr= LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
# OVR METHOD
from sklearn.linear_model import LogisticRegression
```

```python
lr1= LogisticRegression(multi_class="ovr")
lr1.fit(x_train, y_train)
```

```python
lr1.score(x_test, y_test)*100
```

```python
# Multinomial Classification
from sklearn.linear_model import LogisticRegression
lr2= LogisticRegression(multi_class='multinomial')
lr2.fit(x_train, y_train)
```

```python
lr2.score(x_test, y_test)*100
```
