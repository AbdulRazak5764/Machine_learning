```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv(r"C:\Users\Shaik.AbdulRazak\OneDrive\Documents\Raw data\placement.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="cgpa",y="score",data=data,hue="placed")
plt.show()
```

```python
x=data.iloc[:,:-1]
y=data["placed"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
# to know if the data is normal distribution are not
```

```python
sns.kdeplot(data=data["cgpa"])
plt.show()
```

```python
sns.kdeplot(data=data["score"])
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
gnb=GaussianNB()
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test)*100
```

```python
mnb=MultinomialNB()
mnb.fit(x_train,y_train)
```

```python
mnb.score(x_test,y_test)*100
```

```python
bnb=BernoulliNB()
bnb.fit(x_train,y_train)
```

```python
bnb.score(x_test,y_test)*100
```
