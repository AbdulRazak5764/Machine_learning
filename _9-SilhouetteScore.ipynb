# SilhouetteScore

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv(r"C:\Users\Shaik.AbdulRazak\OneDrive\Documents\Raw data\Iris row.csv")
data.head()
```

```python
sns.pairplot(data=data)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(2,21):
    km=KMeans(n_clusters=i,init="k-means++")
    km.fit(data)
    wcss.append(km.inertia_)
```

```python
plt.figure(figsize=(10,5))
plt.plot([i for i in range(2,21)],wcss,marker="o")
plt.xlabel("No of Cluster")
plt.ylabel("wcss")
plt.xticks([i  for i in range(2,21)])
plt.grid(axis="x")
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
kmn=KMeans(n_clusters=3)
data["predict"]=kmn.fit_predict(data)
```

```python
kmn.labels_
```

```python
silhouette_score(data,labels=kmn.labels_)
```

```python
ss=[]
for i in range(2,21):
    km1=KMeans(n_clusters=i)
    km1.fit(data)
    ss.append(silhouette_score(data,km1.labels_))
```

```python
plt.plot( ss, marker="o")
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouttee Score")
plt.show()
```
